#Data Preprocessing

##Finding Redundant columns

```python
df.nunique()
```

###As Claim_number is unique sequence number to all rows so it is not adding any information. Also, Loss_code, Loss_description,TP_type_insd_pass_front,TP_type_pass_multi etc.  are having only one value for more than 96% which lead to no information addition. Droping all these columns will not impact data.

```python
col = df.columns
overfit_col = []
for i in col:
    counts = df[i].value_counts()
    zeros = counts.iloc[0]
    
    if zeros / len(df) * 100 > 96:
        print(zeros,len(df))
        overfit_col.append(i)
overfit_col
```

```python
df=df.drop(overfit_col, axis=1)
```

##Finding Null values in data frame

```python
df.isnull().sum()
```

##Treating the Missing values

```python
df['Weather_conditions']=df.groupby('date_of_loss')['Weather_conditions'].fillna(df['Weather_conditions'].mode()[0])
df['Weather_conditions'].unique()
```
